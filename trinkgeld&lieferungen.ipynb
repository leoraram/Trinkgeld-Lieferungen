Trinkgeld & Lieferungen

Ich arbeite seit einigen Monaten als Rider in der Auslieferung von Lebensmitteln. Der Stundenlohn ist leider relativ gering, durch das Trinkgeld meiner Kund_innen und Boni, die ich pro Lieferung bekomme kann sich mein Nettostundenlohn jedoch ganz gut erhöhen. Schnell fiel mir auf, dass es Tage gab, an denen ich mehr Kund_innen belieferte und somit auch mehr Trinkgeld und Boni bekam, als an anderen. Ich wollte heraus finden, ob es dahinter bestimmte Muster gab, um meine Schichten gegebenenfalls effektiver planen zu können, um also mehr Trinkgeld pro Stunde kassieren bzw. mehr Kund_innen pro Stunde beliefern zu können.

```python

%matplotlib inline
import matplotlib.pyplot as plt

xs_week_tph = [1, 2, 3, 4, 5, 6, 7]
ys_week_tph = []
xs_week_cph = [1, 2, 3, 4, 5, 6, 7]
ys_week_cph = []

mon_t_sum = 0 #Summe des Trinkgeldes aller Montage
mon_h_sum = 0 #Summe der Arbeitsstunden aller Montage
mon_c_sum = 0 #Summe der Kunden/Lieferungen aller Montage

tue_t_sum = 0 
tue_h_sum = 0 
tue_c_sum = 0 

wed_t_sum = 0 
wed_h_sum = 0 
wed_c_sum = 0 

thu_t_sum = 0
thu_h_sum = 0
thu_c_sum = 0

fri_t_sum = 0
fri_h_sum = 0
fri_c_sum = 0

sat_t_sum = 0
sat_h_sum = 0
sat_c_sum = 0

sun_t_sum = 0
sun_h_sum = 0
sun_c_sum = 0

with open("trinkgeld&lieferungen.csv" , "r") as file:
    
    for line in file:
        data = line.strip().split(";")
        
        
        if data[0] == "mon" and int(data[3]) > 0:
            
            mon_t_sum = mon_t_sum + int(data[4])
            mon_h_sum = mon_h_sum + int(data[3])
            mon_c_sum = mon_c_sum + int(data[5])
            
            mon_tph = mon_t_sum / mon_h_sum
            mon_cph = mon_c_sum / mon_h_sum
            
            
        if data[0] == "tue" and int(data[3]) > 0:
            
            tue_t_sum = tue_t_sum + int(data[4])
            tue_h_sum = tue_h_sum + int(data[3])
            tue_c_sum = tue_c_sum + int(data[5])
            
            tue_tph = tue_t_sum / tue_h_sum
            tue_cph = tue_c_sum / tue_h_sum
            
            
        if data[0] == "wed" and int(data[3]) > 0:
            
            wed_t_sum = wed_t_sum + int(data[4])
            wed_h_sum = wed_h_sum + int(data[3])
            wed_c_sum = wed_c_sum + int(data[5])
            
            wed_tph = wed_t_sum / wed_h_sum
            wed_cph = wed_c_sum / wed_h_sum
            
            
        if data[0] == "thu" and int(data[3]) > 0:
            
            thu_t_sum = thu_t_sum + int(data[4])
            thu_h_sum = thu_h_sum + int(data[3])
            thu_c_sum = thu_c_sum + int(data[5])
            
            thu_tph = thu_t_sum / thu_h_sum
            thu_cph = thu_c_sum / thu_h_sum
            
            
        if data[0] == "thu" and int(data[3]) > 0:
            
            thu_t_sum = thu_t_sum + int(data[4])
            thu_h_sum = thu_h_sum + int(data[3])
            thu_c_sum = thu_c_sum + int(data[5])
            
            thu_tph = thu_t_sum / thu_h_sum
            thu_cph = thu_c_sum / thu_h_sum
            
            
        if data[0] == "fri" and int(data[3]) > 0:
            
            fri_t_sum = fri_t_sum + int(data[4])
            fri_h_sum = fri_h_sum + int(data[3])
            fri_c_sum = fri_c_sum + int(data[5])
            
            fri_tph = fri_t_sum / fri_h_sum
            fri_cph = fri_c_sum / fri_h_sum
            
            
        if data[0] == "sat" and int(data[3]) > 0:
            
            sat_t_sum = sat_t_sum + int(data[4])
            sat_h_sum = sat_h_sum + int(data[3])
            sat_c_sum = sat_c_sum + int(data[5])
            
            sat_tph = sat_t_sum / sat_h_sum
            sat_cph = sat_c_sum / sat_h_sum
            
            
        if data[0] == "sun" and int(data[3]) > 0:
            
            sun_t_sum = sun_t_sum + int(data[4])
            sun_h_sum = sun_h_sum + int(data[3])
            sun_c_sum = sun_c_sum + int(data[5])
            
            sun_tph = sun_t_sum / sun_h_sum
            sun_cph = sun_c_sum / sun_h_sum
 

    ys_week_tph.append(mon_tph)
    ys_week_tph.append(tue_tph)
    ys_week_tph.append(wed_tph)
    ys_week_tph.append(thu_tph)
    ys_week_tph.append(fri_tph)
    ys_week_tph.append(sat_tph)
    ys_week_tph.append(sun_tph)
    
    ys_week_cph.append(mon_cph)
    ys_week_cph.append(tue_cph)
    ys_week_cph.append(wed_cph)
    ys_week_cph.append(thu_cph)
    ys_week_cph.append(fri_cph)
    ys_week_cph.append(sat_cph)
    ys_week_cph.append(sun_cph)
                     

print("Mon:")
print("Tip per Hour: " + str(round(mon_tph, 2)))
print("Cus per Hour: " + str(round(mon_cph, 2)))
print()
print()
print("Tue:")
print("Tip per Hour: " + str(round(tue_tph, 2)))
print("Cus per Hour: " + str(round(tue_cph, 2)))
print()
print()
print("Wed:")
print("Tip per Hour: " + str(round(wed_tph, 2)))
print("Cus per Hour: " + str(round(wed_cph, 2)))
print()
print()
print("Thu:")
print("Tip per Hour: " + str(round(thu_tph, 2)))
print("Cus per Hour: " + str(round(thu_cph, 2)))
print()
print()
print("Fri:")
print("Tip per Hour: " + str(round(fri_tph, 2)))
print("Cus per Hour: " + str(round(fri_cph, 2)))
print()
print()
print("Sat:")
print("Tip per Hour: " + str(round(sat_tph, 2)))
print("Cus per Hour: " + str(round(sat_cph, 2)))
print()
print()
print("Sun:")
print("Tip per Hour: " + str(round(sun_tph, 2)))
print("Cus per Hour: " + str(round(sun_cph, 2)))
print()
print()
print()
print()
print("Tip per Hour:")
plt.plot(xs_week_tph, ys_week_tph)
plt.show()
print("1 = Monday")
print("7 = Sunday")
print()
print()
print()
print()
print("Cus per Hour:")
plt.plot(xs_week_cph, ys_week_cph)
plt.show()
print("1 = Monday")
print("7 = Sunday")
            
```
